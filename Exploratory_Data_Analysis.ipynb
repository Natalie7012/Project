```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
plt.title("Highest cost of living")
plt.plot(Cost_Of_Living_df.Cost_of_Living_Index.head(10))
```

```python
Cost_Of_Living_df.sort_values('Cost_of_Living_Index', ascending = True).tail(10).plot.barh(x='Country',
                                                                       y='Cost_of_Living_Index',
                                                                       figsize=(12, 6),
                                                                       title='Top 10 Cost of Living Index by Country',
                                                                       color='red');
```

```python
Cost_Of_Living_df.sort_values('Rent_Index', ascending = True).tail(10).plot.barh(x='Country',
                                                                       y='Rent_Index',
                                                                       figsize=(12, 6),
                                                                       title='Top 10 Rent Index by Country',
                                                                       color='green');
```

```python
Cost_Of_Living_df.sort_values('Groceries_Index', ascending = True).tail(10).plot.barh(x='Country',
                                                                       y='Groceries_Index',
                                                                       figsize=(12, 6),
                                                                       title='Top 10 Groceries_Index by Country',
                                                                       color='Blue');
```

```python

Cost_Of_Living_df.sort_values('Restaurant_Price_Index', ascending = True).tail(10).plot.barh(x='Country',
                                                                       y='Restaurant_Price_Index',
                                                                       figsize=(12, 6),
                                                                       title='Top 10 Restaurant_Price_Index by Country',
                                                                       color='Blue');
```
